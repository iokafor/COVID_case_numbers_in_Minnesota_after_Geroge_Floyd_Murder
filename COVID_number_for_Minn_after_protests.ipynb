```
# import libraries

import pandas as pd
import numpy as np

#import viz

import plotly.express as px
```

# import data

```
# import data

#csse_covid_19_daily_reports_us_05-25-20 through 6-01-20

df_5_25_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/05-25-2020.csv")
df_5_26_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/05-26-2020.csv")
df_5_27_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/05-27-2020.csv")
df_5_28_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/05-28-2020.csv")
df_5_29_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/05-29-2020.csv")
df_5_30_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/05-30-2020.csv")
df_6_01_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/06-01-2020.csv")


#csse_covid_19_daily_reports_us_6-9 through 6-15

df_6_09_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/06-09-2020.csv")
df_6_10_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/06-10-2020.csv")
df_6_11_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/06-11-2020.csv")
df_6_12_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/06-12-2020.csv")
df_6_13_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/06-13-2020.csv")
df_6_14_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/06-14-2020.csv")
df_6_15_20 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/06-15-2020.csv")


```

```
df_5_25_20.head(1)
                
```

```
df_5_25_20_Minn = df_5_25_20[df_5_25_20["Province_State"]== "Minnesota"]
```

```
df_5_25_20_Minn.head(1)
```

```
list_of_df_Covid_before_murder = [df_5_26_20,df_5_25_20,df_5_27_20,df_5_28_20,df_5_29_20,df_5_30_20,df_6_01_20]
list_of_df_Covid_after_murder = [df_6_09_20,df_6_10_20,df_6_11_20,df_6_12_20,df_6_13_20,df_6_14_20,df_6_15_20]
```

```
df_Covid_before_murder = pd.concat(list_of_df_Covid_before_murder)
df_Covid_after_murder = pd.concat(list_of_df_Covid_after_murder) 
```

```
df_Covid_before_murder_Minn = df_Covid_before_murder[df_Covid_before_murder["Province_State"] == "Minnesota"]
df_Covid_after_murder_Minn = df_Covid_after_murder[df_Covid_after_murder["Province_State"] == "Minnesota"]
```

```
df_Covid_after_murder_Minn.head(10)
                                
```

```
df_Covid_before_murder_Minn.head(10)
```

```
# Calculating the confirmed to people tested ratio before and after killing
df_Covid_before_murder_Minn["confirmed to people tested ratio"] = df_Covid_before_murder_Minn["Confirmed"]/ df_Covid_before_murder_Minn["People_Tested"]
df_Covid_after_murder_Minn["confirmed to people tested ratio"] = df_Covid_after_murder_Minn["Confirmed"] / df_Covid_after_murder_Minn["People_Tested"]
```

```
df_Covid_after_murder_Minn.head(10)
```

```
df_Covid_before_murder_Minn.head(10)
```

```
df_Covid_before_after_murder_Minn_confirmed_to_tested_ratio = [df_Covid_after_murder_Minn,df_Covid_before_murder_Minn]
```

```
df_Covid_before_after_murder_Minn_confirmed_to_tested_ratio = pd.concat(df_Covid_before_after_murder_Minn_confirmed_to_tested_ratio)
```

```
df_Covid_before_after_murder_Minn_confirmed_to_tested_ratio.head(10)
```

```
df_Covid_before_after_murder_Minn_confirmed_to_tested_ratio.sort_values(by='Last_Update')
```

# Line graph of confirmed to tested over time

```
Scatter_Covid_ratio_in_Minn = px.scatter(df_Covid_before_after_murder_Minn_confirmed_to_tested_ratio, x = "Last_Update", y = "confirmed to people tested ratio", title = "Confirmed COVID-19 cases to number of people tested ratio from 5/26/20 - 6/16/20 in Minnesota")
```

```
Scatter_Covid_indident_rate_in_Minn = px.scatter(df_Covid_before_after_murder_Minn_confirmed_to_tested_ratio, x = "Last_Update" , y = "Incident_Rate", title = "COVID-19 incident rate from 5/26/20 - 6/16/20 in Minnesota")
```
